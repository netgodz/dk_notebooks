# 制作排序直方图

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.family'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

%matplotlib inline
```

```python
y = np.array([0.02, 0.021, 0.032, 0.041, 0.055, 0.060, 0.072, 0.08, 0.078, 0.07, 0.061, 0.052, 0.042, 0.034, 0.025, 0.022])*100
x = np.arange(y.size)
x_str = [str(ix+1) for ix in x]
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

# Save the chart so we can loop through the bars below.
bars = ax.bar(x=x, height=y, tick_label=x_str)

# Axis formatting.
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)
#ax.spines['left'].set_visible(False)
#ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_ylim([0, 15])

# Add text annotations to the top of the bars.
#

# Add labels and a title. Note the use of `labelpad` and `pad` to add some
# extra space between the text and the tick labels.
ax.set_xlabel('观测数据的秩(Order)', labelpad=15, color='#333333', fontsize=20)
ax.set_ylabel('观测出现频率(%)', labelpad=15, color='#333333', fontsize=20)
ax.set_title('排序直方图', pad=15, color='#333333',
             weight='bold', fontsize=20)

fig.tight_layout()
```
